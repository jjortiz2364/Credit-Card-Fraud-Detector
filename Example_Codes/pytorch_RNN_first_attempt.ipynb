## Test with Actual Data

### Splitting Data

```python
decision_features = df.columns[:-1]
target_feature = df.columns[-1]
```

```python
X = df[decision_features]
y = df[target_feature]
```

```python
#example_tensor = Tensor(X)
```

```python
rnn = RNN(input_size=len(X.columns), hidden_size=2*len(X.columns), nonlinearity='relu')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=999)
```

```python
X_train_values, y_train_values = X_train.to_numpy(), y_train.to_numpy()
X_test_values, y_test_values = X_test.to_numpy(), y_test.to_numpy()
```

```python
X_train_tensor, y_train_tensor = tensor(X_train_values), tensor(y_train_values, dtype=torch.int64)
```

```python
X_test_tensor, y_test_tensor = tensor(X_test_values), tensor(y_test_values, dtype=torch.int64)
```
